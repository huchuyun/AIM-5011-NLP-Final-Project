## Process object data
### （1）Extract the object names in all json files and write them to a local folder

```python
import os
import pandas as pd
import re
root_dir=r"E:\object"
output_dir=r"E:\object3"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
# Read the json files in the object folder one by one
for file in os.listdir(root_dir):
    file_name=root_dir+'\\'+file
    output_name=output_dir+'\\'+file
    pattern=re.compile('  name: "(\w+)"')
    print(file,":",end='')
    obj=open(file_name,'r')
    outputfile=open(output_name,'w')
    # Find the name line in the json file
    for line in obj:
        result=pattern.match(line)
        if result:
            # Write the content after name to the txt file
            outputfile.write(result.group(1))
            outputfile.write(',')
            print(result.group(1),end=',')
    outputfile.close()
    print(end='\n')
    obj.close()
```

### （2）Label theimage and objects according to the file name in the 'hate' and 'nothate' folders and write it into a csv file

```python
import numpy as np
root_dir=r"E:\object3"
output_dir_1=r"E:\object_name\hate_objectname"
output_dir_0=r"E:\object_name\nothate_objectname"
if not os.path.exists(output_dir_0):
    os.makedirs(output_dir_0)
if not os.path.exists(output_dir_1):
    os.makedirs(output_dir_1)
    
# Save the names of all files in the hate folder
hate=os.listdir(r'E:\Hate')
# Save the names of all files in the nothate folder
nothate=os.listdir(r'E:\NotHate')

result=pd.DataFrame(columns=('index','image_name','object','label'))
# a is the number of files that have be read, and also the index of the dataframe
a=0
# Read the json files in the object folder one by one
for file in os.listdir(root_dir):
    print(file)
    file_name=root_dir+'\\'+file
    output_name_0=output_dir_0+'\\'+file[:-4]+'txt'
    output_name_1=output_dir_1+'\\'+file[:-4]+'txt'
    obj=open(file_name,'r')
    # If the name of the image is in the hate folder
    if file[:-5] in hate:
        outputfile=open(output_name_1,'w')
        # If the file is not empty
        if os.path.getsize(file_name):
            # Write the file to the txt file in the hate folder and generate a record in the dataframe
            for line in obj:
                outputfile.write(line[:-1])
                result=result.append(pd.DataFrame({'index':[a],'image_name':[file[:-5]],'object':[line[:-1].split(",")],'label':[1]}))
        # If the file is empty
        else:
            # generate a null record in the dataframe
            result=pd.concat([result,pd.DataFrame({'index':[a],'image_name':[file[:-5]],'object':[[]],'label':[1]})])
        print(a)
        a+=1
        outputfile.close()
    # If the name of the image is in the nothate folder
    elif file[:-5] in nothate:
        outputfile=open(output_name_0,'w')
        # If the file is not empty
        if os.path.getsize(file_name):
            # Write the file to the txt file in the nothate folder and generate a record in the dataframe
            for line in obj:
                outputfile.write(line[:-1])
                result=result.append(pd.DataFrame({'index':[a],'image_name':[file[:-5]],'object':[line[:-1].split(",")],'label':[0]}))
        # If the file is empty
        else:
            # generate a null record in the dataframe
            result=pd.concat([result,pd.DataFrame({'index':[a],'image_name':[file[:-5]],'object':[[]],'label':[0]})])
        print(a)
        a=a+1
        outputfile.close()
    # If the name of the image is not in the hate folder and not in the nothate folder
    else:
        print(a,"both not")
    obj.close()
```

```python
# Set index column as index
result=result.set_index(['index'])
result
```

```python
# Set index column as index
result.to_csv(r"E:\object_name.csv",index=False)
```
